# Temperature change at the Phoenix SKY airport

The plot below shows the temperature change, in Celsius, between the daily average temperature and a historical average temperature. 
The data, downloaded from the National Oceanic and Atmospheric Administration (NOAA), ftp://ftp.ncdc.noaa.gov/pub/data/ghcn/v3, contains the monthly average temperature collected since 1880s until now. 

The dataset contains information from hundreds of weather stations, the plot below corresponds the weather station closest to Phoenix's SKY airport. 

The average monthly temperature was calculated from the data between 1951 and 1980. 
The line shown is the result of a linear fit, which we could use to predict the raise in temperature in the future. It has a slope of 0.45 Celsius (0.81 Farenheit), every 10 years.
